```python
A=df.resample('MS').count()
```

```python
from matplotlib.pyplot import rcParams
rcParams['figure.figsize'] = 18, 6 
A.FIRE_SIZE.plot()
```

```python
pd.plotting.lag_plot(A.FIRE_SIZE,lag=1)
```

```python
#Transforming the values 
A['log_FIRE_SIZE']=np.log(A.FIRE_SIZE)
A.log_FIRE_SIZE.plot()
```

```python
#Making mean constant 
A['log_FIRE_SIZE_diff']=A.log_FIRE_SIZE.diff(1)
A[['log_FIRE_SIZE','log_FIRE_SIZE_diff']].head(24)
A.log_FIRE_SIZE_diff.plot()
```

```python
#Seasonal Adjustement 
A['stationary']=A.log_FIRE_SIZE_diff.diff(periods=12)
A[['log_FIRE_SIZE','log_FIRE_SIZE_diff','stationary']]
A.stationary.plot()
```

```python
A['shift_1']=A.stationary.shift(1)
#df[['df.stationary','df.shift_1']].head()


plt.scatter(A.shift_1,A.stationary)
cor=A.stationary.corr(A.shift_1)
pd.plotting.lag_plot(A.stationary,lag=1)
cor
```

```python
import statsmodels.api as sm

A.stationary.dropna(inplace=True)

sm.graphics.tsa.plot_acf(A.stationary,lags=30, zero=False)
plt.show()
```

```python
sm.graphics.tsa.plot_pacf(A.stationary,lags=30, zero=False)
plt.show()
```

```python
from statsmodels.tsa.ar_model import AR 
from statsmodels.tsa.arima_model import ARMA # ARMA for Moving Average model
from statsmodels.tsa.arima_model import ARIMA
X=A.dropna()
t=-12

train=X[:t]
test=X[t:]
model=AR(train.stationary,freq='MS')
#model=ARIMA(train.stationary,freq='MS',order = (0,1,1))

#model=ARIMA(train.stationary,order=(0,1,0))
model_fitted=model.fit(maxlag=15)
#nos of previous value 
print("Lag",model_fitted.k_ar)
#params-coef for each lag phi value
print("Coef values",model_fitted.params)
```

```python
predictions=model_fitted.predict(start=len(train),end=len(X)-1)
predicted=pd.Series(predictions,index=A[t:].index,name='predicted')
#predicted=pd.Series(predictions,name='predicted')
plt.title('Prediction Vs Actual sales before transformation', fontsize = 16)
plt.xlabel('Date', fontsize = 16)
test.stationary.plot()
predicted.plot()
plt.xticks(fontsize = 14)
plt.legend()
plt.show()
```

```python
pred=[np.exp(predicted[0])*train.FIRE_SIZE[-1]* train.FIRE_SIZE[-12]/train.FIRE_SIZE[-13]]

for i in range(1,12):
    pred.append(np.exp(predicted[i])*pred[-1] * train.FIRE_SIZE[i-12]/train.FIRE_SIZE[i-13])

pred_final=pd.Series(pred,index=A[t:].index,name='predicted')
pd.concat([test.FIRE_SIZE,pred_final],axis=1)

plt.title('Prediction Vs Actual sales after transformation', fontsize = 16)
plt.xlabel('Date', fontsize = 16)
plt.ylabel('Total', fontsize = 16)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
#plt.ylim(0, 2)
A.FIRE_SIZE[-36:].plot(color='gray',marker='o')
pred_final.plot(marker='o')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

RMSE=np.sqrt(mean_squared_error(pred_final,A.FIRE_SIZE[t:]))
print("The RMSE value is", RMSE.round(1))
```

```python
from pandas import read_csv
from matplotlib import pyplot
from statsmodels.tsa.ar_model import AR
from sklearn.metrics import mean_squared_error
#series = read_csv('daily-minimum-temperatures.csv', header=0, index_col=0)
# split dataset
X = A.FIRE_SIZE
train, test = X[1:len(X)-12], X[len(X)-12:]
# train autoregression
model = AR(train)
model_fit = model.fit()
window = model_fit.k_ar
coef = model_fit.params
# walk forward over time steps in test
history = train[len(train)-window:]
history = [history[i] for i in range(len(history))]
predictions = list()
for t in range(len(test)):
	length = len(history)
	lag = [history[i] for i in range(length-window,length)]
	yhat = coef[0]
	for d in range(window):
		yhat += coef[d+1] * lag[window-d-1]
	obs = test[t]
	predictions.append(yhat)
	history.append(obs)
	print('predicted=%f, expected=%f' % (yhat, obs))
error = np.sqrt(mean_squared_error(test, predictions))
print('Test RMSE: %.3f' % error)

```
